### **Paper 7: Artificial Intelligence Application in Diagnosing, Classifying, Localizing, Detecting and Estimation the Severity of Skin Condition in Aesthetic Medicine: A Review**

DOI: 10.1002/der2.70015

**Content summary**

This paper is a  review of artificial intelligence and machine learning applications in aesthetic dermatology. It focuses on how AI is used for diagnosing, classifying, localizing, detecting, and assessing the severity of skin conditions. AI systems can analyze medical images and patient data with high precision, for more accurate diagnoses, personalized treatments, and improved clinical efficiency. It also shows the benefits of using AI tools in this field, such as reduced human error and lower healthcare costs. Limitations also occcur, with challenges like algorithmic bias, data privacy, and the need for high-quality datasets.

**Dataset used**

Since this is a review paper, it does not rely on a single dataset but summarizes multiple studies using diverse datasets from Public medical databases like MEDLINE, PubMed and Ovid, which were used to collect the reviewed studies. 

Most datasets consist of annotated skin images for conditions like acne, psoriasis, skin cancer, and aging signs: 
- Clinical and dermoscopic image datasets --> ranging from about 100 to over 500,000 images
- Patient datasets --> from about 1000 patients in diagnostic systems
- Specialized datasets --> CT/MRI scans, selfie images, scalp/hair images

**Models used**

The review covers a wide range of AI/ML models. CNN-based models are the most dominant for image-based diagnosis and classification tasks: 
- Deep Learning models: CNNs, DCNNs, Mask R-CNN, AlexNet, VGG16, GANs
- Traditional ML models: SVM, Random Forest, Naive Bayes, KNN
- Hybrid approaches: CNN + SVM, CNN + Random Forest
- Specialized techniques: YOLO for object detection, GAFCM clustering, Fourier-based ANN


**Performance**

Across studies, AI systems show high performance. Typical accuracy ranges: 85% to 96% depending on the task.
- Some Scalp classification systems models outperform dermatologists in consistency and accuracy
- Deep learning models generally outperform traditional ML methods --> CNN vs SVM/random forest
- There is also a high sensitivity reported for disease detection (often over 90%)

Pros:
- Improved diagnostic accuracy and speed
- Effective severity assessment and monitoring
- Strong performance in image-based classification and detection tasks
- Comparable or superior results to human experts in some cases

limitations:
- Dataset bias and lack of diversity
- Small sample sizes in some studies
- Need for better generalization and regulatory validation

In conclusion, the paper concludes that AI has strong potential to transform dermatology, but further work is needed for clinical deployment.

### **Paper 8: Computational enzyme design by catalytic motif scaffolding**

DOI: 10.1038/s41586-025-09747-9

**Content summary**

This paper presents Riff-Diff, as a novel computational pipeline for de novo enzyme design using a combination of machine learning and atomistic modeling. The method uses scaffolding catalytic motifs, which are specific arrangements of amino acids responsible for catalysis, into newly designed protein backbones.  Where traditional approaches  require extensive screening and optimization, Riff-Diff aims to generate functional enzymes in a “one-shot” design process. The method is validated by designing enzymes for two reactions: the retro-aldol reaction and the Morita–Baylis–Hillman (MBH) reaction.

**Dataset used**

The study relys on a combination of datasets:
- Protein Data Bank (PDB) structures for catalytic motifs and structural templates
- Artificial motif libraries --> generated computationally from catalytic residue configurations
- Experimental datasets from designed enzyme variants --> 35 retro-aldolase designs, 60 MBH enzyme designs
- Structural and biochemical validation data -->  SAXS, crystallography, kinetic assays
- External datasets like PDBbind were used to model realistic binding pocket properties

**Models used**

The Riff-Diff pipeline integrates multiple computational models and tools. This hybrid approach combines diffusion-based generative modeling, deep learning, and physics-based methods into a unified enzyme design framework.

Used models: 
- RFdiffusion for protein backbone generation
- LigandMPNN for sequence design
- Rosetta (FastRelax, CoupledMoves) for structural refinement
- AlphaFold2 / AlphaFold3 and ESMFold for structure prediction and validation
- Molecular dynamics simulations for analyzing flexibility and catalytic behavior


**Performance**

Riff-Diff achieves a one-shot enzyme design, with a high success rate.  


Retro-aldolases showed 91% successfull catalytic activity:
- Catalytic efficiency: Best designs achieved
> kcat ≈ 3 × 10⁻² s⁻¹ and  ~10⁶-fold rate acceleration over uncatalyzed reactions
- This is Comparable to evolved enzymes: Performance rivals enzymes optimized through directed evolution, but with far fewer designs --> tens vs. thousands

MBH reaction: 
- Designed enzymes outperform earlier computational designs and even some evolved variants
- High stability: Many designs remain stable above 90 °C
- High stereoselectivity: Up to 99% enantiomeric excess
 
Limitations of Riff-Diff: 
- Activity prediction remains difficult --> poor correlation with metrics like RMSD
- Enzyme flexibility and dynamics still limit performance
- Requires predefined catalytic motifs --> challenge for entirely new reactions

 Traditional computational enzyme design show low activity and requires extensive screening, where as Riff-Diff has a higher initial activity with minimal screening, approaching evolved enzyme performance. In conclusion, the method represents a major step toward practical, scalable enzyme design without iterative optimization cycles.

### **Paper 9: Language model-guided anticipation and discovery of mammalian metabolites**

DOI: 10.1038/s41586-025-09969-x

**Content summary**

This paper introduces a chemical language model designed to predict and discover previously unknown mammalian metabolites. The model is called DeepMet and it learns patterns from known metabolite structures and generates new, plausible metabolite candidates, addressing the problem of “metabolic dark matter, which are defined as unidentified signals in metabolomics data. By integrating DeepMet with mass spectrometry (MS/MS) workflows, the study demonstrates both computational anticipation and experimental validation of new metabolites, significantly expanding current metabolic knowledge.

**Dataset used**

The study combines multiple datasets:
- Training data: ~2,046 known human metabolites represented as SMILES strings
- Pretraining data: Drug-like molecules from the ChEMBL database
Evaluation datasets:
- Human Metabolome Database (HMDB), including newly added metabolites
- Large-scale metabolomics datasets -->  29.1 million MS/MS spectra from human blood samples
- Experimental datasets from 23 mouse tissues and biofluids
Additional validation via synthesized chemical standards and LC–MS/MS experiments


These datasets enable both in silico validation and experimental confirmation of predictions.

**Models used**

The framework integrates several models:
- DeepMet: an LSTM-based chemical language model trained on SMILES representations of metabolites
- Random Forest classifiers: for evaluation, prioritization, and meta-learning
- CFM-ID: for predicting MS/MS spectra and matching experimental data
- Additional statistical and similarity metrics --> Tanimoto similarity, Fréchet ChemNet distance

Key innovation is using sampling frequency from the language model as a confidence measure for how likely a molecule exists biologically.

**Performance**

DeepMet shows strong performance in metabolite prediction and discovery:


Generation capability:
- Recovered 81% of newly added HMDB metabolites
- Generated ~90% of held-out metabolites
Structure prediction (mass only):
- ~29% top-1 accuracy for exact structure identification --> Correct structure often within top-ranked candidates


With MS/MS integration:
- ~50–52% top-1 accuracy for structure identification
- Improved ranking and structural similarity vs baseline methods


With meta-learning (combined features):
- Accuracy increases to ~70%


Experimental validation:
- Dozens of new metabolites discovered (e.g., 36 total, including 16 validated in mouse tissues)


In comparison to baselines, DeepMet Outperforms simpler methods like AddCarbon and improves metabolite annotation coverage compared to HMDB-only approaches.

Key outcomes:
>Enables prediction of metabolites before experimental discovery, and thus expands metabolite annotation in large datasets. It also bridges AI-generated hypotheses with wet-lab validation

Limitations:
>Limited to metabolite-like chemical space (bias toward known structures). it has also difficulty distinguishing isomers using MS data alone. Requires experimental validation for confirmation and performance depends on similarity to training data.


Overall, the paper demonstrates that language models can meaningfully expand biochemical knowledge, not just analyze existing data.
