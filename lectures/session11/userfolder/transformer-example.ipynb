# Use a pretrained Transformer to predict SMILES

Load model: GPT2LMHeadModel - trained for predicting smiles


```python
from transformers import GPT2LMHeadModel, PreTrainedTokenizerFast

# Load tokenizer & pretrained model
tokenizer = PreTrainedTokenizerFast.from_pretrained(
    "jonghyunlee/MolGPT_pretrained-by-ZINC15"
)

tokenizer.pad_token = "<pad>"
tokenizer.bos_token = "<bos>"
tokenizer.eos_token = "<eos>"

model = GPT2LMHeadModel.from_pretrained("jonghyunlee/MolGPT_pretrained-by-ZINC15")
model.eval()

def generate_smiles(n=5, temperature=1.0):
    outputs = model.generate(
        max_length=128,
        num_return_sequences=n,
        do_sample=True,
        temperature=temperature,
        bos_token_id=tokenizer.bos_token_id,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.pad_token_id
    )
    return [tokenizer.decode(o, skip_special_tokens=True) for o in outputs]

# Test generation
print("Generated SMILES:")
for smi in generate_smiles(5):
    print(smi)
```
