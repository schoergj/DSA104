```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
```

```python
true_values = [10,10,10,10,0]
predicted_values = [10,10,10,10,100]
```

```python
mean_squared_error(true_values,predicted_values)
```

```python
np.sqrt(mean_squared_error(true_values,predicted_values))
```

```python
mean_absolute_error(true_values,predicted_values)
```
